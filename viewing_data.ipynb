```python
# importing data 

import pandas as pd
houses = pd.read_csv('House Price India.csv')
houses.head()
```

```python
houses.describe()
```

```python
# cleaning data

# we see that there is a house with 33 bedrooms! but the living area is small. assume this is anomaly 
houses.sort_values(by = 'No of bedrooms', ascending = False)
print(houses['No of bedrooms'].max())
print(houses['No of bathrooms'].max())

# removing the outlier 
mask = houses['No of bedrooms'] == 33
houses = houses[~mask]

# check it has been successfully removed 
houses['No of bedrooms'].describe()
```

```python
# want to rename the columns for better consistency and shorten them 
d = {'Date':'date',
     'No of bedrooms':'bedrooms',
     'No of bathrooms':'bathrooms',
     'living area':'living_area',
     'lot area':'lot_area',
     'No of floors':'floors',
     'waterfront present':'waterfront',
     'No of views':'views',
     'house condition':'condition',
     'house grade':'grade',
     'house area(excluding basement)':'house_area',
     'Area of the basement':'basement_area',
     'Built Year':'year_built',
     'Renovation Year':'year_renovated',
     'Postal Code':'post_code',
     'Lattitude':'lattitude',
     'Longitude':'longitude',
     'No of schools nearby':'schools',
     'Distance from the airport':'airport_distance',
     'Price':'price'
    }

houses = houses.rename(d, axis = "columns")

houses.head()
```

```python
#save the clean file to a csv 

houses.to_csv('HousePricesClean.csv', index = False)
```

```python
# analysis : how does lot area affect price?

print(houses[['lot_area', 'price']].corr(method = 'pearson'))

# we can see that based on just a single factor, there is no significant correlation 
```

```python
# does num bedrooms scale with house area?

houses[['house_area', 'bedrooms']].corr(method = 'pearson')
```

```python
# split data into test and train
size = len(houses)

split_index = int(size * 0.75)

train = houses.iloc[:split_index]
test = houses.iloc[split_index:].reset_index()

train.to_csv('train.csv')
test.to_csv('test.csv')
```
